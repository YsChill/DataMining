```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons, load_iris
from sklearn.decomposition import PCA
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

#Generate a 2D dataset with non-convex clusters


```python
X, y = make_moons(n_samples=300, noise=0.05, random_state=42)

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Visualize
plt.scatter(X_scaled[:, 0], X_scaled[:, 1])
plt.title("Dataset for DBSCAN")
plt.show()
```

#Applying DBSCAN

```python

dbscan = DBSCAN(eps=0.3, min_samples=5)
labels = dbscan.fit_predict(X_scaled)

# Plot clusters
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels, cmap='plasma')
plt.title("DBSCAN Clustering")
plt.show()
```

#Use a real dataset:
* Try load_iris() from sklearn.datasets.
* Apply DBSCAN

```python
iris = load_iris()
X_iris = iris.data
X_iris_scaled = StandardScaler().fit_transform(X_iris)

dbscan_iris = DBSCAN(eps=0.5, min_samples=5)
labels_iris = dbscan_iris.fit_predict(X_iris_scaled)

pca = PCA(n_components=2)
X_2d = pca.fit_transform(X_iris_scaled)

plt.figure(figsize=(8, 5))
plt.scatter(X_2d[:, 0], X_2d[:, 1], c=labels_iris, cmap='plasma')
plt.title("DBSCAN on Iris Dataset (PCA-reduced)")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.colorbar(label='Cluster Label')
plt.show()
```
